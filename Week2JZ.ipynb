```python
#Task 1 - Accept Variable length and Print
def func1(*args):
    for arg in args:
        print(arg)

# Example usage
func1(10, 20, 30, "Hi there", [14, 15, 16])
func1("See me in the future")
```

```python
#Task 2 - Recursive addition
def sum_numbers_recursive(n):
    if n == 0:
        return 0
    else:
        return n + sum_numbers_recursive(n - 1)

# Calculate the sum of numbers from 0 to 15
result = sum_numbers_recursive(15)
print("The sum of numbers from 0 to 15 is:", result)
```

```python
#Task 3a
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Reading the dataset
df = pd.read_csv("company_sales_data.csv")
# df.head()

# Data
x = df['month_number']
y = df['total_profit']

# Plotting the line plot
plt.plot(x, y)

# Adding labels and title
plt.xlabel('Month')
plt.ylabel('Total Profit')
plt.title('Total Profit by Month')

# Display the plot
plt.show()
```

```python
#Task 3b
x = df['month_number']
y1 = df['bathingsoap']
y2 = df['facewash']

# Create subplots
fig, axs = plt.subplots(2)  # 2 rows of subplots

# First subplot (upper)
axs[0].plot(x, y1, color='blue')
axs[0].set_title('Bathing Soap')

# Second subplot (lower)
axs[1].plot(x, y2, color='red')
axs[1].set_title('Face Wash')

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```
